```python
import pandas as pd
from ast import literal_eval
```

```python
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', 100)
```

```python
file = '0-100k.csv'
```

```python
df = pd.read_csv(f'../data/preprocessed/{file}', converters={'tokens': literal_eval, 'nltk_pos': literal_eval, 'spacy_pos': literal_eval,'spacy_label': literal_eval,})
df.head()
```

```python
df['is_person_spacy'] = df['spacy_label'].apply(lambda x: 1 if 'PERSON' in x else 0)
```

```python
df['spacy_pos_len'] = df['spacy_pos'].apply(len)
```

```python
len(df)
```

```python
df_inventor = df[df['is_person_spacy'] == 1]
df_inventor[['title']].head(20)
```

```python
df_non_inventor = df[df['is_person_spacy'] == 0]
```

```python
df_prpn = df_non_inventor[(df_non_inventor['is_prpn_spacy'] == 1) & (df_non_inventor['spacy_pos_len'] <= 2)]

df_non_prpn_1 = df_non_inventor[(df_non_inventor['is_prpn_spacy'] == 1) & (df_non_inventor['spacy_pos_len'] > 2)]
df_non_prpn_2 = df_non_inventor[df_non_inventor['is_prpn_spacy'] == 0]

df_non_prpn = pd.concat([df_non_prpn_1, df_non_prpn_2], ignore_index=True)
```

```python
len(df) == (len(df_inventor) + len(df_prpn) + len(df_non_prpn))
```

```python
len(df)
```

```python
df_non_prpn[['patent_number', 'title', 'spacy_pos']].head(100)
```

```python
len(df) 
```

```python
df_dirty = pd.concat([df_inventor, df_prpn], ignore_index=True)
df_dirty
```

```python
non_terminating_pos = ['ADP', 'CONJ', 'CCONJ', 'DET', 'ADP']

df_non_prpn['bad_endings'] = df['spacy_pos'].apply(lambda x: 1 if x[-1] in non_terminating_pos else 0)
```

```python
df_bad_endings = df_non_prpn[df_non_prpn['bad_endings'] == 1]
df_non_bad_endings = df_non_prpn[df_non_prpn['bad_endings'] == 0]
```

```python
len(df) == len(df_inventor) + len(df_prpn) + len(df_bad_endings) + len(df_non_bad_endings)
```

```python
df_non_bad_endings.head()
```

```python
df_single_pos = df_non_bad_endings[df_non_bad_endings['spacy_pos_len'] == 1]

df_non_single_pos = df_non_bad_endings[df_non_bad_endings['spacy_pos_len'] != 1]
```

```python
len(df) == len(df_inventor) + len(df_prpn) + len(df_bad_endings) + len(df_single_pos) + len(df_non_single_pos)
```

```python
df_non_single_pos['bad_starts'] = df_non_single_pos['spacy_pos'].apply(lambda x: 1 if x[0] in non_terminating_pos else 0)

df_bad_starts = df_non_single_pos[df_non_single_pos['bad_starts'] == 1]
df_non_bad_starts = df_non_single_pos[df_non_single_pos['bad_starts'] == 0]
```

```python
len(df) == len(df_inventor) + len(df_prpn) + len(df_bad_endings) + len(df_single_pos) + len(df_bad_starts) + len(df_non_bad_starts)
```

```python
len(df_non_bad_starts)
```

```python
df_non_bad_starts[['title', 'spacy_pos']].head(100)
```

```python
df_non_bad_starts['starts_with_improve'] = df_non_bad_starts['title'].apply(lambda x: 1 if x.lower().startswith('improve') else 0)
```

```python
df_non_bad_starts[df_non_bad_starts['starts_with_improve'] == 0][['title', 'spacy_pos']].head(603)
```

### Spell Checker

```python
from spellchecker import SpellChecker

spell = SpellChecker()
a = 'Imprbvemenf in spring shackle'
a.split()
```

```python
# find those words that may be misspelled
misspelled = spell.unknown(a.split())
misspelled
```

```python
df_non_bad_starts['title'] = df_non_bad_starts['title'].apply(lambda x: x.replace('-', ' '))
df_non_bad_starts['misspelt_words'] = df_non_bad_starts['title'].apply(lambda x: spell.unknown(x.split()))
df_non_bad_starts['misspelt'] = df_non_bad_starts['misspelt_words'].apply(lambda x: 1 if len(x) > 0 else 0)
```

```python
df_spelling_errors = df_non_bad_starts[df_non_bad_starts['misspelt'] != 0]
df_non_spelling_errors = df_non_bad_starts[df_non_bad_starts['misspelt'] == 0]
```

```python
len(df) == len(df_inventor) + len(df_prpn) + len(df_bad_endings) + len(df_single_pos) + len(df_bad_starts) + len(df_spelling_errors) + len(df_non_spelling_errors)
```

```python
df_non_spelling_errors[['title', 'misspelt_words']].head(100)
```

```python
df_inventor.to_csv(f'../data/cleaned/{file}_inventor.csv', index=False)

df_prpn.to_csv(f'../data/cleaned/{file}_proper_nouns.csv', index=False)

df_bad_endings.to_csv(f'../data/cleaned/{file}_incomplete.csv', index=False)

df_single_pos.to_csv(f'../data/cleaned/{file}_single_pos.csv', index=False)

df_bad_starts.to_csv(f'../data/cleaned/{file}_bad_starts.csv', index=False)

df_spelling_errors.to_csv(f'../data/cleaned/{file}_non_bad_starts.csv', index=False)
df_non_spelling_errors.to_csv(f'../data/cleaned/{file}_non_bad_starts.csv', index=False)
```
