## Patent Title Extraction:

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tqdm import tqdm
```

```python
pd.set_option('display.max_colwidth', None)
```

```python
base_url = 'https://patents.google.com/patent/US'
```

```python
patents = []
titles = []
dates = []

for i in tqdm(range(100_000, 200_000)):
    try:
        url = f'{base_url}{i}'
        # get url instance
        reqs = requests.get(url)
        soup = BeautifulSoup(reqs.text, 'html.parser')
        text = soup.find('title').get_text().split(' - Google Patents')[0]
        patent = text.split(' - ')[0]
        title = text.split(' - ')[1]
        date = soup.time.attrs['datetime']
        patents.append(patent)
        titles.append(title)
        dates.append(date)
    except:
        continue
```

```python
df = pd.DataFrame(list(zip(patents, dates, titles)), columns =['patent_number', 'date', 'title'])
df['date'] =  pd.to_datetime(df['date'])
df.head(20)
```

```python
df.to_csv('first_10k_patents.csv', index=False)
```
